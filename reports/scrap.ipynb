```python
import requests
import urllib.request
import time
import random
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup
```

```python
show = 'the office'
eps = []
seasons = 9
```

```python
for x in range(1, seasons + 1):
    url = 'https://www.imdb.com/title/tt0386676/episodes?season=' + str(x)
    response = requests.get(url)
    s = BeautifulSoup(response.text, 'html.parser')
    tags = s.findAll('strong')
    temp = []
    
    for x in tags:
        tag = str(x)
        
        if 'title' in tag:
            tag = tag[59:]
            tag = tag.split('"')
            temp.append(tag[0])
            
    random.shuffle(temp)
    random.shuffle(temp)
    random.shuffle(temp)
    eps.append(temp)
```

```python
print(show + '\n')
for x in eps:
    for y in x:
        print(y)
    print('\n')
    
```
