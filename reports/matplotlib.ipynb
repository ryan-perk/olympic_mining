## subplots

subplots allow you to have multiple plots in a single figure. the plot is divided in rectangular panes that are numbered rowwise in the format (# rows, # cols, index) where 'index' starts in the top left and moves right, then down.  

_a few parameters.._

- *args: 3-digit integer or csv to describe the subplot  
- projection (optional): the projection type, projections:
> {None, 'aitoff', 'hammer', 'lambert',   
    'mollweide', 'polar', 'rectilinear', str}   
- polar (optional): boolean, if True, the same as projection = 'polar'
- sharex, sharey (optional):  share an axis with another x or y
- label: str, name of the returned plot

https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.subplot.html

```python
# 1 - 3-digit integer
# sub = plt.subplot(221)

# 2 - more generally, csv format
sub = plt.subplot(5, 5, 1)

# adds a subplot with no frame
no_frame = plt.subplot(5, 5, 3, frameon = False)

# adds a aitoff subplot
plt.subplot(5, 5, 5, projection = 'aitoff')

# adds a hammer subplot
plt.subplot(5, 5, 11, projection = 'hammer')

# adds a lambert subplot
plt.subplot(5, 5, 13, projection = 'lambert')

# adds a mollweide subplot
plt.subplot(5, 5, 15, projection = 'mollweide')

# adds a polar subplot
plt.subplot(5, 5, 21, polar = True)

# adds a rectilinear subplot
plt.subplot(5, 5, 23, projection = 'rectilinear')

# adds a red subplot that shares the x-axis with sub
plt.subplot(5, 5, 25, sharex = sub, facecolor = 'red')

# deletes no_frame from the figure
plt.delaxes(no_frame)

# adds no_frame to the figure again
plt.subplot(no_frame)
```
